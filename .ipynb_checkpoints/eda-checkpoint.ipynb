```python
# EXPLORATORY DATA ANALYSIS (EDA) WITH PYTHON

# EDA is basically the first look at your data
# Getting to know your data before you do anything serious with it

# Identifying patterns within a dataset
# Understanding the relationships between features
# Identifying Outliers 
# Identifying mistakes within the dataset - missing values / Null values / Placeholder 

# There are hundred of ways to perform EDA, and we can't show you all
# Focus: Cleaning, descriptive statistics, and visualization. 
```

```python
# Step 1: Import our libraries
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
print('Libraries loaded successfully')
```

```python
# Step 2: load our dataset - World population dataset
path = 'world_population.csv' 
df = pd.read_csv(path);
print('Dataset loaded successfully')
```

```python
# Step 3: Understand your dataset - High level information about your dataset
df
```

```python
# change decimal point to 2 point
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df
```

```python
# see top 5 entries
df.head()
```

```python
# see last 5 entries
df.tail()
```

```python
# info about dataset
df.info()
```

```python
# Statistical overview of dataset
df.describe()
```

```python
#Handling Null values/Missing values/ placeholders
df.isnull().sum()
```

```python
# Unique value
df.nunique()
```

```python
#top rank countries
df.sort_values(by="World Population Percentage", ascending=False).head(10) #by default sort in ascending order
```

```python
# Correlation - Numeric values
df.select_dtypes(include='number').corr()
```

```python
# Visualize - heatmap
sns.heatmap(df.select_dtypes(include='number').corr(), annot = True)

# increase size
plt.rcParams["figure.figsize"] = (10, 7)
plt.show()
```
